```python
import matplotlib.pyplot as plt
import numpy as np
import math

Nsample = 10
degree = 3
_lamda = 0   
_mu = 0
_sigma = 0.1

featureX = 2*np.pi*np.random.random(size=Nsample)

noise = np.random.normal(_mu, _sigma, Nsample)
featureY = np.array([np.sin(featureX[i])+noise[i] for i in range(Nsample)])

matrixX = featureX**0
for i in range(1,degree+1):
    matrixX = np.vstack((matrixX, featureX**i))

coef = np.dot( np.dot( featureY, matrixX.T ),   np.linalg.inv( np.dot(matrixX, matrixX.T ) +  _lamda * np.eye(degree + 1) ) )

X = 2*np.pi*np.random.random(size=100)
X.sort()
S = np.sin(X)
predict = []
for x in X:
    result = 0
    for i in range(degree+1):
        result += coef[i] * x**i
    predict.append(result)
plt.ylim(-1.5,1.5)
plt.plot(X, S, color="green",linewidth=3.0)
plt.plot(X, predict, color="red",linewidth=3.0)
plt.scatter(featureX, featureY, 50,color ='blue',marker = 'o')
plt.show()
```
